```python
df = df.drop('new_column', axis=1)
#drop the created column. Have to specifiy axis. Rows have axis = 0. Columns have axis = 1.
```

```python
df = df.reset_index(drop=True)
#after removing nulls this function resets the row index counter 
df.index = df.index + 1
#start index from 1 rather than 0
```

```python
df.head()
```

```python
df.loc[:,['release_year','rating']]
#return every row with release year and rating
```

```python
df.loc[df['release_year']>2000]
#returns everything released after 2000
```

```python
df.iloc[0,1]
#finds the exact cell value
```

```python
df = df[df['rating']=='R']
#filters to show all movies rated R
```
